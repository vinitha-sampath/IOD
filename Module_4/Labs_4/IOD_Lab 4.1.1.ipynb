#### 2.6  Predict Using Linear Regression

[Simple Linear Regression](https://en.wikipedia.org/wiki/Simple_linear_regression#Fitting_the_regression_line)

### $$\hat{y} = \beta_0 + \beta_1 x$$

> ### $$ \beta_1 = \frac{\sum_{i=1}^n (y_i - \bar{y} ) (x_i - \bar{x} )}{\sum_{i=1}^n (x_i - \bar{x})^2} $$

and

> ### $$ \beta_0 = \bar{y} - \beta_1\bar{x} $$

Given variables:
- $\bar{y}$ : the sample mean of observed values $Y$
- $\bar{x}$ : the sample mean of observed values $X$

##### 2.7 Define X, Y

- X = Predictor Variable
- Y = Target Variable

```python
# ANSWER
X = df['sq__ft'] #Predictor Variable
Y = df['price'] #Target Variable
```

##### 2.7.A Calculate beta_1

 ### $$ \beta_1 = \frac{\sum_{i=1}^n (y_i - \bar{y} ) (x_i - \bar{x} )}{\sum_{i=1}^n (x_i - \bar{x})^2} $$

```python
# ANSWER
def calculate_beta_1(X,Y):
    Xbar = X.mean()
    Ybar = Y.mean()
    a = np.sum((X-Xbar)*(Y-Ybar))
    b = np.sum(np.square(X-Xbar))
    beta_1 = a/b
    return beta_1
beta_1 = calculate_beta_1(X, Y)
beta_1
```

##### 2.7.B Calculate beta_0

### $$ \beta_0 = \bar{y} - \beta_1\bar{x} $$

```python
# ANSWER
def calculaye_beta_0(X,Y):
    Xbar = X.mean()
    Ybar = Y.mean()
    beta_0 = Ybar - (beta_1*Xbar)
    return beta_0
beta_0 = calculaye_beta_0(X,Y)
beta_0
```

##### 2.7.C Find Yhat (Predicted Value Using Beta 0 and Beta 1)

### $$\hat{y} = \beta_0 + \beta_1 x$$

```python
Yhat = beta_0 + (beta_1*X)
print(Yhat)
```

```python
# ANSWER
def calculate_yhat(X, beta_0, beta_1):
    yhat = beta_0 + (beta_1*X)
    return yhat
Yhat =  calculate_yhat(X, beta_0, beta_1)
Yhat
```

##### 2.7.D Find Residuals

### $$ \text{residual}_i = y_i - \hat{y}_i$$

```python
# ANSWER
def calculate_resi(Y, Yhat):
    residual = Y - Yhat
    resi = abs(residual)
    return resi 
resi = calculate_resi(Y, Yhat)
resi
```

##### 2.7.E Find R-Squared

The coefficient $R^2$ is defined as $$ R^2 = 1 - \frac{SS_{res}}{SS_{tot}} $$, where $$ SS_{res} $$ is the residual sum of squares $$ SS_{res} = \sum_{i=1}^n \left(y_i - \hat{y}_i\right)^2 $$ and $$ SS_{res} $$ is the total sum of squares $$ SS_{tot} = \sum_{i=1}^n \left(y_i - \bar{y}\right)^2 $$

The best possible score is 1.0 and it can be negative (because the model can be arbitrarily worse).

**Write a function to calculate R- Square value.**

The total sum of squares:

### $$ SS_{tot} = \sum_{i=1}^n \left(y_i - \bar{y}\right)^2 $$

The residual sum of squares:

### $$ SS_{res} = \sum_{i=1}^n \left(y_i - \hat{y}_i\right)^2 $$

$R^2$:

### $$ R^2 = 1 - \frac{SS_{res}}{SS_{tot}} $$

```python
# ANSWER
def calculate_total_sum_of_square(Y):
    Ybar = np.mean(Y)
    SStot = np.square(Y-Ybar)
    SStot = np.sum(SStot)
    return SStot
SStot = calculate_total_sum_of_square(Y)
SStot
```

```python
def calculate_residual_sum_of_squares(Y, Yhat):
    SSres = np.square(Y - Yhat)
    SSres = np.sum(SSres)
    return SSres
SSres = calculate_residual_sum_of_squares(Y, Yhat)
SSres
```

```python
def calculate_r_square(Y, Yhat):
    SStot = calculate_total_sum_of_square(Y)
    SSres = calculate_residual_sum_of_squares(Y, Yhat)
    Rsquare = 1 - (SSres/SStot)
    return Rsquare
Rsquare = calculate_r_square(Y, Yhat)
Rsquare
```
